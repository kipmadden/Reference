```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
from empiricaldist import Pmf
```

```python
pmf = Pmf.from_seq([1, 2, 2, 4, 5], name='example')
```

```python
pmf.plot()
plt.legend();
```

```python
pmf.head()
```

```python
pmf.bar()
plt.legend();
```

```python
cdf = pmf.make_cdf()
cdf.plot()
plt.legend();
```

```python
cdf.step()
plt.legend();
```

```python
surv = cdf.make_surv()
surv.plot()
plt.legend();
```

```python
surv = cdf.make_surv()
surv.step()
plt.legend();
```

```python
haz = surv.make_hazard()
haz.plot()
plt.legend();
```

```python
haz.bar()
plt.legend();
```
